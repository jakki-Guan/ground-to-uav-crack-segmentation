# Final Report Assets Preview

This notebook is a lightweight preview layer for the frozen paper-facing assets under `results/report_assets/final_comparison`.

It does five things:
1. locates the repo root robustly from the current working directory,
2. optionally reruns `make_final_report_assets.py`,
3. renders the summary plus the main comparison tables,
4. previews the supervision-scaling data that drives the final figure,
5. displays the generated PNG assets and the selected qualitative cases.


```python
from pathlib import Path
import subprocess
import sys

import pandas as pd
from IPython.display import Image, Markdown, display

REPO_CANDIDATES = [Path.cwd(), Path.cwd().parent]
REPO_ROOT = next(
    path.resolve()
    for path in REPO_CANDIDATES
    if (path / "make_final_report_assets.py").exists()
)
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

ASSET_DIR = REPO_ROOT / "results" / "report_assets" / "final_comparison"
MAIN_RESULTS_MD = ASSET_DIR / "main_results.md"
MAIN_RESULTS_CSV = ASSET_DIR / "main_results.csv"
COMPARISON_ONLY_MD = ASSET_DIR / "comparison_only_variants.md"
COMPARISON_ONLY_CSV = ASSET_DIR / "comparison_only_variants.csv"
SUPERVISION_CSV = ASSET_DIR / "supervision_scaling.csv"
SUPERVISION_PNG = ASSET_DIR / "supervision_scaling.png"
QUAL_CASES_CSV = ASSET_DIR / "qualitative_cases.csv"
QUAL_PNG = ASSET_DIR / "qualitative_progression_test.png"
SUMMARY_MD = ASSET_DIR / "summary.md"

REBUILD_ASSETS = False
if REBUILD_ASSETS or not MAIN_RESULTS_CSV.exists():
    subprocess.run([sys.executable, str(REPO_ROOT / "make_final_report_assets.py")], check=True)

print(f"Repo root: {REPO_ROOT}")
print(f"Asset dir: {ASSET_DIR}")
print("Available files:")
for path in sorted(ASSET_DIR.iterdir()):
    print(f"- {path.name}")
```

```python
display(Markdown(SUMMARY_MD.read_text(encoding="utf-8")))

main_df = pd.read_csv(MAIN_RESULTS_CSV)
comparison_df = pd.read_csv(COMPARISON_ONLY_CSV)
qual_df = pd.read_csv(QUAL_CASES_CSV)
supervision_df = pd.read_csv(SUPERVISION_CSV)

display(Markdown("## Main Table"))
display(main_df)

display(Markdown("## Comparison-Only Operating Points"))
display(comparison_df)
```

```python
display(Markdown("## B2 Supervision Data"))
display(supervision_df)

display(Markdown("## Stage-Wise Pivot (IoU)"))
iou_pivot = main_df.pivot(index=["stage", "setting", "labeled_samples"], columns="model", values="iou")
display(iou_pivot)

display(Markdown("## Stage-Wise Pivot (F1)"))
f1_pivot = main_df.pivot(index=["stage", "setting", "labeled_samples"], columns="model", values="f1")
display(f1_pivot)
```

```python
display(Markdown("## Supervision-Scaling Figure"))
display(Image(filename=str(SUPERVISION_PNG)))

display(Markdown("## Qualitative Progression Figure"))
display(Image(filename=str(QUAL_PNG)))
```

```python
display(Markdown("## Selected Qualitative Cases"))
display(qual_df)

display(Markdown(MAIN_RESULTS_MD.read_text(encoding="utf-8")))
display(Markdown(COMPARISON_ONLY_MD.read_text(encoding="utf-8")))
```
